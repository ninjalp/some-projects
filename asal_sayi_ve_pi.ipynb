#  asal sayı  tespit programı 

```python
import math

number = int(input("Enter a number: "))
is_prime = True
if number <= 1:
    is_prime = False
else:
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

# pi sayısı rakamları hazır kütüp

```python
from mpmath import mp


mp.dps = 150528
pi_value = mp.pi
print(pi_value)
type(pi_value)
```

```python
pi_str=str(pi_value)
```

```python
pi_list=[int(digit) for digit in pi_str if digit.isdigit()]
```

```python
pi_list
print(pi_list)
print(len(pi_list))
```

```python
(h,w,c) = pi_image.shape
```

```python
bakalim=pi_image.reshape(h*w,c)
```
